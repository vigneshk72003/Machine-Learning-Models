# Hierarchical Clustering

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
dataset=pd.read_csv("Mall_Customers.csv")
x=dataset.iloc[:,3:5].values
```

## Using the dendrogram to find the optimal number of cluster

```
from scipy.cluster import hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method="ward"))#using dendrogram keyword from sch class
#iam ammplying variance method(ward) for each coloumn in x which will be helping to find the euclidean distance
#at each datapoints

#we cant give the title which in dataset overhere, this dendrogram, kmean all predict the no. of optimal clusters
#through this clusters only we will get the hirearichal,kmean cluster graphs
plt.title("hirearichal clusteres(dendrogram)")
plt.xlabel("customers")#that takes the datapoints
plt.ylabel("euclidean distance")
plt.show()
```

from this we will predict no.of clusters(see in note), we got both 3 & 5 clusters,here using 5 clusters

## Training the Hierarchical Clustering model on the dataset(now we need the graph in cluster format, we done with the agglomerative herarichal cluster)

```
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")#linkage is to same use same variance(ward)method for this
#chnged cluster also
#the parameters gets shown while writing the class() properly, affinity tells the calculation method for
#y-axis
y_hc=ac.fit_predict(x)
```

```
print(y_hc)#since only 5 clusters code shows from index 0 to 4
```

this conveys that which customers belongs to which clusters based upon the coloumns in x

## Visualising the clusters

```
#based on the clusters from 1 to 5(index:0 to 4)we plotted each clusters on its own size, colour, and label
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c="red",label="cluster 1")
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c="blue",label="cluster 2")
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,c="black",label="cluster 3")
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,c="orange",label="cluster 4")
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,c="yellow",label="cluster 5")
plt.title("annual income vs spending(by agglomerativ hirearchical cluster)")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()#this will says the label with colors
plt.show()
```
