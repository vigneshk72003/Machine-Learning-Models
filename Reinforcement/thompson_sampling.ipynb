# Thompson Sampling

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
dataset=pd.read_csv("Ads_CTR_Optimisation.csv")
```

## Implementing Thompson Sampling

```
import random #here we will be using bayes reward distriubution
N=10000#rows
d=10#columns
#based on the given algorithm,intially no ads will be rewarded so making each has 0
ads_selected=[]
numbers_of_rewards_0=[0]*d
numbers_of_rewards_1=[0]*d
total_reward=0
for i in range(0,N):
  #before moving in column in each round, we need necssary terms mentioned in algorithm
  ad=0
  max_random=0
  for j in range(0,d):
    #using beta from random class
    theta_i=random.betavariate(numbers_of_rewards_0[j]+1,numbers_of_rewards_1[j]+1)
    if(theta_i > max_random):
      max_random=theta_i
      ad=j
  #no need of else part, because in previous algorithm we put if based on the denominator
  #we have to implement next part within this for loop,becuase ads choosed in column forma
  ads_selected.append(ad)  ##final value
  reward=dataset.values[i,ad]
  if(reward==1):
    numbers_of_rewards_1[ad]=numbers_of_rewards_1[ad]+1
  else:
    numbers_of_rewards_0[ad]=numbers_of_rewards_0[ad]+1
  total_reward=total_reward+reward#for this no need of index,because at each reward
    #index is plotted and added each time



```

## Visualising the results - Histogram

```
plt.hist(ads_selected)
plt.title("Ads CTR optimisation(thompson sampleing)")
plt.xlabel("ads")
plt.ylabel("no. of times each ads selected by the user")
plt.show()
```
